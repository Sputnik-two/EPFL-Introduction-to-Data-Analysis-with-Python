#### Task 1: Compute the probability that a randomly selected commune with elevation over 2000 is from the canton of Valais.

```python
above_2000 = (
    df.groupby("Canton")["Elevation"]
    .apply(lambda x: (x > 2000).sum())
    .reset_index(name="no. of communes")
)
above_2000
```

```python
Sum_of_communes = sum(above_2000["no. of communes"])
Sum_of_communes
```

```python
VS_above_2000 = df[(df.Elevation >= 2000) & (df.Canton == "VS")].Canton.value_counts(
    normalize=False
)
VS_above_2000
```

__Consideration__

There are a total of 96 communes with an elevation over 2000. 96 communes are over 2000 meter. 49 of them are in Valais.
Therefore 49 out of 96 communes fullfill the conditions: So the probability is 49/96 = 0.51

```python
VS_above_2000 / Sum_of_communes
```

#### Task 1: Define a data frame matrix whose rows correspond to communes and the columns to the cantons. Fill in the matrix with 0/1 values where entry (i,j) is a 1 if the commune in row i is in the canton in column j and a 0 otherwise.

```python
df.iloc[1111]["Commune"]  # to check the results in the end
```

```python
df.iloc[1111]["Canton"]  # to check the results in the end
```

```python
my_cantons = df["Canton"].unique()  # a list of all the cantons
my_cantons
```

```python
matrix = np.zeros((2191, 26))  # a matrix with 2191 rows and 26 columns
matrix
```

```python
for i in range(2191):  # for-loop for the rows
    for j in range(26):  # for-loop for the columns
        if df.iloc[i]["Canton"] == my_cantons[j]:
            matrix[i, j] = 1  # if the commune is in the canton print 1
```

```python
my_result_1 = pd.DataFrame(data=matrix, index=df["Commune"], columns=my_cantons)
my_result_1.iloc[1111:1112, :]
```

```python
# Same result, much faster solution
my_result_2 = pd.crosstab(df["Commune"], df["Canton"])
```
